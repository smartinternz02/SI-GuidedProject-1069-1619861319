```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.subplot(2,2,4)
plt.plot(y,x,'r')
plt.subplot(2,2,2)
plt.scatter(x,y,color ='orange')
plt.subplot(2,2,3)
plt.scatter(y,x,color = 'pink')
```

```python
fig = plt.figure()
axes = fig.add_axes([1,1,5,5])
axes2 = fig.add_axes([1,2,2,3])
axes.plot(x,y,'b')
axes2.scatter(x,y,color = 'red')
```
